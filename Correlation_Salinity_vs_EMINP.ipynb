```python
import matplotlib.pyplot as plt

seas = ["Dead Sea", "Red Sea", "Arabian Sea (N)", "Arabian Sea (S)", "BoB (N)", "BoB (S)"]
salinities = [347, 41, 37, 36, 35, 33]
e_p_values = [3150, 2100, 1800, 600, 0, -500]
correlation_coefficients = [-0.99, -0.85, -0.75, -0.55, 0.00, 0.70]

plt.figure(figsize=(12, 8))
plt.scatter(e_p_values, salinities, c=correlation_coefficients, cmap="coolwarm")

plt.xlabel("Evaporation - Precipitation (mm/year)")
plt.ylabel("Salinity (ppt)")
plt.title("Correlation Between Salinity and E-P in Different Seas")

sm = plt.cm.ScalarMappable(cmap="coolwarm", norm=plt.Normalize(min(correlation_coefficients), max(correlation_coefficients)))
sm.set_array([])
plt.colorbar(sm, label="Correlation Coefficient (r)")

for i, sea in enumerate(seas):
    plt.annotate(sea, (e_p_values[i], salinities[i]), textcoords="offset points", xytext=(0, 10))

plt.grid(True)
plt.tight_layout()
plt.savefig(r'C:\Users\pavan\Downloads\Earth System\S_vs_Ep_Corr.'+'jpg')
plt.show()
```
